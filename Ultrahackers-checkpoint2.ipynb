# Data Cleaning

First we imported the data using pandas. We noticed that the data is missing headers so we added labels for each column. 

```python
import pandas as pd
import numpy as np

numOfHeaders = 29

#set the labels for the features
headers = ["feature "+str(i) for i in range(1,numOfHeaders)]

# mark the first column to be the target
headers = ["output"] + headers

#import the data and set the columns' headers to the labels
data = pd.read_csv("HIGGS_train.csv",names = headers, low_memory=False)
data
```

We noticed that there are some duplicated rows that we dropped down.

```python
numOfDuplicatedRows = len(np.where(data.duplicated() == True)[0])
print("The number of duplicated rows = ",numOfDuplicatedRows)

data = data.drop_duplicates()
data
```

We searched for rows with missing data, and we got their indices and dropped them accordingly. 

```python
print("The number of rows that have null values is : ", len(np.where(data.isna().any(axis=1) == True)[0]))
#get the row with the missing data
rows = data[data.isna().any(axis=1)]
#get the index of that row and drop it
data = data.drop(rows.index,axis=0)
data
```

We noticed that the data is not of the same type. We got the indices of the columns and the rows where the data is of type string. we looped over the cells with string data and converted them to float if applicable. if not, we replaced them with the mode of the column.

```python
print("Is the data of the same type?",data.dtypes.nunique() == 1)
#get the indices of the columns that have non float values 
cols = np.where(data.dtypes != "float64")
#get the headers of those rows
cols = np.array([headers[i] for i in cols[0]])
#create a copy of the data
copyOfData = data
#convert everything to numeric type, if a data cell is non numeric, it will be converted to nan
copyOfData = copyOfData.apply(pd.to_numeric,errors='coerce')
#indices of the rows that contains non numeric values 
indices = np.where(copyOfData.isna().any(axis = 1))[0]

#loop over the rows
for row in indices:
    #loop over the elements of the row
    for col in cols:
        #get the element at that row at that col
        num = data.iloc[row][col]
        try: 
            #if applicable, get the numeric value of the string
            data = data.replace({col:{num:float(num.strip('"'))}})
        except:
            #replace the field with the mode
            data = data.replace({col:{num:data[col].mode()[0]}}) 
data
```

# Data Training:

# 1) Training the data with the logistic regression model:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y = np.array(data.iloc[:,0])
X = np.array(data.iloc[:,1:])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model= LogisticRegression()
model.fit(X_train, y_train)
print("The accuracy of the logistic regression model = ",model.score(X_test,y_test))
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import regularizers
tf.random.set_seed(42)
X_train_1 = np.array(X_train[:,0:21])
X_train_2 = np.array(X_train[:,21:28])
X_train_1 = X_train_1.astype(np.float32)
X_train_1 = tf.convert_to_tensor(X_train_1)

X_train_2 = X_train_2.astype(np.float32)
X_train_2 = tf.convert_to_tensor(X_train_2)

X_test_1 = np.array(X_test[:,0:21])
X_test_2 = np.array(X_test[:,21:28])

X_test_1 = X_test_1.astype(np.float32)
X_test_1 = tf.convert_to_tensor(X_test_1)

X_test_2 = X_test_2.astype(np.float32)
X_test_2 = tf.convert_to_tensor(X_test_2)
early_stop = EarlyStopping(monitor='val_loss', patience=15)
model1 = tf.keras.models.Sequential([
tf.keras.layers.Dense(4000, activation='relu', kernel_regularizer=regularizers.l2(0.0009)),
tf.keras.layers.Dense(2000, activation='relu'),
tf.keras.layers.Dense(900, activation='relu'),
tf.keras.layers.Dense(200)
])
model2 = tf.keras.models.Sequential([
tf.keras.layers.Dense(3000, activation='relu'),
tf.keras.layers.Dense(1000, activation='relu'),
tf.keras.layers.Dense(500, activation='relu'),
tf.keras.layers.Dense(200 )
])
# create the user input and point to the base network
input1 = tf.keras.layers.Input(shape=(21))
vu = model1(input1)
vu = tf.linalg.l2_normalize(vu, axis=1)
# create the item input and point to the base network
input2 = tf.keras.layers.Input(shape=(7))
vm = model2(input2)
vm = tf.linalg.l2_normalize(vm, axis=1)
# measure the similarity of the two vector outputs
output = tf.keras.layers.Dot(axes=1)([vu, vm])
# specify the inputs and output of the model
model = tf.keras.models.Model([input1, input2], output)
# Specify the cost function
#cost_fn = tf.keras.losses.binary_crossentropy()

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit([X_train_1, X_train_2], y_train, epochs=100,batch_size= 200, validation_data=([X_test_1,X_test_2], y_test), callbacks=[early_stop])

loss, accuracy = model.evaluate([X_test_1,X_test_2], y_test)
loss2, accuracy2 = model.evaluate([X_train_1,X_train_2], y_train)
# Print the test loss and accuracy
print('Test accuracy:', accuracy)
print('Train accuracy:', accuracy2)

# Epoch 1/100
# 2397/2397 [==============================] - 29s 11ms/step - loss: 0.6409 - accuracy: 0.6455 - val_loss: 0.5814 - val_accuracy: 0.6931
# Epoch 2/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5815 - accuracy: 0.6928 - val_loss: 0.5824 - val_accuracy: 0.6873
# Epoch 3/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.6530 - accuracy: 0.6117 - val_loss: 0.6114 - val_accuracy: 0.6715
# Epoch 4/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5745 - accuracy: 0.7000 - val_loss: 0.5543 - val_accuracy: 0.7153
# Epoch 5/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5596 - accuracy: 0.7101 - val_loss: 0.5491 - val_accuracy: 0.7176
# Epoch 6/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5510 - accuracy: 0.7157 - val_loss: 0.5464 - val_accuracy: 0.7222
# Epoch 7/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5524 - accuracy: 0.7146 - val_loss: 0.5416 - val_accuracy: 0.7235
# Epoch 8/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5466 - accuracy: 0.7194 - val_loss: 0.5477 - val_accuracy: 0.7216
# Epoch 9/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5394 - accuracy: 0.7248 - val_loss: 0.5302 - val_accuracy: 0.7326
# Epoch 10/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5397 - accuracy: 0.7246 - val_loss: 0.5444 - val_accuracy: 0.7236
# Epoch 11/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5452 - accuracy: 0.7215 - val_loss: 0.5290 - val_accuracy: 0.7318
# Epoch 12/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5335 - accuracy: 0.7295 - val_loss: 0.5238 - val_accuracy: 0.7380
# Epoch 13/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5265 - accuracy: 0.7347 - val_loss: 0.5333 - val_accuracy: 0.7297
# Epoch 14/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5247 - accuracy: 0.7367 - val_loss: 0.5203 - val_accuracy: 0.7399
# Epoch 15/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5210 - accuracy: 0.7392 - val_loss: 0.5190 - val_accuracy: 0.7405
# Epoch 16/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5263 - accuracy: 0.7347 - val_loss: 0.5187 - val_accuracy: 0.7410
# Epoch 17/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5519 - accuracy: 0.7172 - val_loss: 0.5452 - val_accuracy: 0.7215
# Epoch 18/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5359 - accuracy: 0.7290 - val_loss: 0.5258 - val_accuracy: 0.7364
# Epoch 19/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5399 - accuracy: 0.7248 - val_loss: 0.5518 - val_accuracy: 0.7166
# Epoch 20/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5370 - accuracy: 0.7286 - val_loss: 0.5250 - val_accuracy: 0.7378
# Epoch 21/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5311 - accuracy: 0.7328 - val_loss: 0.5199 - val_accuracy: 0.7419
# Epoch 22/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5246 - accuracy: 0.7376 - val_loss: 0.5284 - val_accuracy: 0.7343
# Epoch 23/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5268 - accuracy: 0.7356 - val_loss: 0.5442 - val_accuracy: 0.7255
# Epoch 24/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5345 - accuracy: 0.7303 - val_loss: 0.5384 - val_accuracy: 0.7274
# Epoch 25/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5327 - accuracy: 0.7317 - val_loss: 0.5204 - val_accuracy: 0.7418
# Epoch 26/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5263 - accuracy: 0.7371 - val_loss: 0.5161 - val_accuracy: 0.7449
# Epoch 27/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5298 - accuracy: 0.7344 - val_loss: 0.5264 - val_accuracy: 0.7381
# Epoch 28/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5374 - accuracy: 0.7276 - val_loss: 0.5294 - val_accuracy: 0.7342
# Epoch 29/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5197 - accuracy: 0.7417 - val_loss: 0.5127 - val_accuracy: 0.7464
# Epoch 30/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5377 - accuracy: 0.7271 - val_loss: 0.5328 - val_accuracy: 0.7315
# Epoch 31/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5220 - accuracy: 0.7405 - val_loss: 0.5175 - val_accuracy: 0.7456
# Epoch 32/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5179 - accuracy: 0.7429 - val_loss: 0.6614 - val_accuracy: 0.5950
# Epoch 33/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5274 - accuracy: 0.7365 - val_loss: 0.5139 - val_accuracy: 0.7464
# Epoch 34/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5120 - accuracy: 0.7463 - val_loss: 0.5104 - val_accuracy: 0.7480
# Epoch 35/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5304 - accuracy: 0.7332 - val_loss: 0.5412 - val_accuracy: 0.7292
# Epoch 36/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5194 - accuracy: 0.7418 - val_loss: 0.5132 - val_accuracy: 0.7467
# Epoch 37/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5126 - accuracy: 0.7463 - val_loss: 0.5116 - val_accuracy: 0.7475
# Epoch 38/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5122 - accuracy: 0.7464 - val_loss: 0.5123 - val_accuracy: 0.7476
# Epoch 39/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5066 - accuracy: 0.7504 - val_loss: 0.5129 - val_accuracy: 0.7466
# Epoch 40/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5227 - accuracy: 0.7396 - val_loss: 0.5107 - val_accuracy: 0.7481
# Epoch 41/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5151 - accuracy: 0.7440 - val_loss: 0.5145 - val_accuracy: 0.7466
# Epoch 42/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5230 - accuracy: 0.7401 - val_loss: 0.5179 - val_accuracy: 0.7441
# Epoch 43/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5255 - accuracy: 0.7385 - val_loss: 0.5138 - val_accuracy: 0.7465
# Epoch 44/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5138 - accuracy: 0.7454 - val_loss: 0.5149 - val_accuracy: 0.7451
# Epoch 45/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5089 - accuracy: 0.7495 - val_loss: 0.5106 - val_accuracy: 0.7502
# Epoch 46/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5033 - accuracy: 0.7525 - val_loss: 0.5147 - val_accuracy: 0.7514
# Epoch 47/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5075 - accuracy: 0.7500 - val_loss: 0.5099 - val_accuracy: 0.7507
# Epoch 48/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5014 - accuracy: 0.7544 - val_loss: 0.5094 - val_accuracy: 0.7506
# Epoch 49/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5163 - accuracy: 0.7437 - val_loss: 0.5090 - val_accuracy: 0.7508
# Epoch 50/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5054 - accuracy: 0.7510 - val_loss: 0.5065 - val_accuracy: 0.7516
# Epoch 51/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5136 - accuracy: 0.7455 - val_loss: 0.5214 - val_accuracy: 0.7403
# Epoch 52/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5048 - accuracy: 0.7515 - val_loss: 0.5091 - val_accuracy: 0.7514
# Epoch 53/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4974 - accuracy: 0.7567 - val_loss: 0.5117 - val_accuracy: 0.7521
# Epoch 54/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4970 - accuracy: 0.7569 - val_loss: 0.5160 - val_accuracy: 0.7448
# Epoch 55/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5003 - accuracy: 0.7549 - val_loss: 0.5090 - val_accuracy: 0.7523
# Epoch 56/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5134 - accuracy: 0.7459 - val_loss: 0.5227 - val_accuracy: 0.7412
# Epoch 57/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5109 - accuracy: 0.7490 - val_loss: 0.5190 - val_accuracy: 0.7457
# Epoch 58/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5158 - accuracy: 0.7458 - val_loss: 0.5192 - val_accuracy: 0.7451
# Epoch 59/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5065 - accuracy: 0.7526 - val_loss: 0.5167 - val_accuracy: 0.7500
# Epoch 60/100
# 2397/2397 [==============================] - 25s 10ms/step - loss: 0.5025 - accuracy: 0.7544 - val_loss: 0.5109 - val_accuracy: 0.7511
# Epoch 61/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4970 - accuracy: 0.7580 - val_loss: 0.5084 - val_accuracy: 0.7525
# Epoch 62/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4948 - accuracy: 0.7591 - val_loss: 0.5094 - val_accuracy: 0.7536
# Epoch 63/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4995 - accuracy: 0.7558 - val_loss: 0.5111 - val_accuracy: 0.7529
# Epoch 64/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.5014 - accuracy: 0.7538 - val_loss: 0.5155 - val_accuracy: 0.7484
# Epoch 65/100
# 2397/2397 [==============================] - 24s 10ms/step - loss: 0.4902 - accuracy: 0.7626 - val_loss: 0.5094 - val_accuracy: 0.7509
# 3745/3745 [==============================] - 12s 3ms/step - loss: 0.5094 - accuracy: 0.7509
# 14980/14980 [==============================] - 48s 3ms/step - loss: 0.4860 - accuracy: 0.7642
# Test accuracy: 0.7509012222290039
# Train accuracy: 0.7642383575439453
```
